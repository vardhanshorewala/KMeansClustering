```
def variance(data):
  variance = (statistics.stdev(data))**2
  return variance
```

```
def wrong(dist, x, y, coll, cha_x, cha_y):
  wrong = []

  mean = sum(dist)/len(dist)
  threshold = mean + 1.85*statistics.stdev(dist)

  for i in range(len(dist)):
    if dist[i] >= threshold:
      wrong.append(dist[i])
    
  for a in wrong:
    del dist[dist.index(a)]
    curr1 = coll.index(a)
    cha_x.append(x[curr1])
    cha_y.append(y[curr1])
    del x[curr1]
    del y[curr1]
    del coll[curr1]
```

```
def averagenew(arr, new):
  sum = 0
  for val in arr:
    sum += val
  mean = sum/len(arr)
  new.append(mean)
```

```
curr_variance = []
curr_boldmin = []
curr_carl = []
curr_sil = []
curr_anom = []

for rar in range(0, 10):  

  x=random.randint(1000, size=(1000))
  y=random.randint(1000, size=(1000))

  loop_thresh = 50

  variance_arr = []
  sil_arr = []
  bold_arr = []
  carl_arr = []
  change_x =[]
  change_y =[]


  for value in range(loop_thresh): 

    x = list(x)
    y = list(y)

    K = 10

    X = np.array(list(zip(x, y)))

    kmeans_model = KMeans(n_clusters=K).fit(X)

    arr0 = []
    arr1 = []
    arr2 = []
    arr3 = []
    arr4 = []
    arr5 = []
    arr6 = []
    arr7 = []
    arr8 = []
    arr9 = []

    dist0 = []
    dist1 = []
    dist2 = []
    dist3 = []
    dist4 = []
    dist5 = []
    dist6 = []
    dist7 = []
    dist8 = []
    dist9 = []

    collated_distances = []

    for i in range(len(kmeans_model.labels_)):
      curr = [x[i], y[i]]
      if kmeans_model.labels_[i] == 0:
        arr0.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[0], curr)
        dist0.append(a)
      if kmeans_model.labels_[i] == 1:
        arr1.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[1], curr)
        dist1.append(a)
      if kmeans_model.labels_[i] == 2:
        arr2.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[2], curr)
        dist2.append(a)
      if kmeans_model.labels_[i] == 3:
        arr3.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[3], curr)
        dist3.append(a)    
      if kmeans_model.labels_[i] == 4:
        arr4.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[4], curr)
        dist4.append(a) 
      if kmeans_model.labels_[i] == 5:
        arr5.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[5], curr)
        dist5.append(a)
      if kmeans_model.labels_[i] == 6:
        arr6.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[6], curr)
        dist6.append(a) 
      if kmeans_model.labels_[i] == 7:
        arr7.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[7], curr)
        dist7.append(a)      
      if kmeans_model.labels_[i] == 8:
        arr8.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[8], curr)
        dist8.append(a)   
      if kmeans_model.labels_[i] == 9:
        arr8.append([x[i], y[i]])
        a = distance.euclidean(kmeans_model.cluster_centers_[8], curr)
        dist8.append(a)       
      collated_distances.append(a)

    conc = []

    for i in range(len(x)):
      conc.append([x[i], y[i]])

    curr_df = pd.DataFrame(conc, columns = list(['x', 'y'])) 
    print(len(x))
    new_var = (variance(dist0) + variance(dist1) + variance(dist2) + variance(dist3) + variance(dist4)+ variance(dist5)+ variance(dist6) + variance(dist7) + variance(dist8) + variance(dist9))/10
    new_sil = silhouette_score(curr_df, kmeans_model.labels_)
    new_bold = davies_bouldin_score(curr_df, kmeans_model.labels_)
    new_carl = calinski_harabasz_score(curr_df, kmeans_model.labels_)

    variance_arr.append(new_var)
    sil_arr.append(new_sil)
    carl_arr.append(new_carl)
    bold_arr.append(new_bold)

    wrong(dist0, x, y, collated_distances, change_x, change_y)
    wrong(dist1, x, y, collated_distances, change_x, change_y)
    wrong(dist2, x, y, collated_distances, change_x, change_y)
    wrong(dist3, x, y, collated_distances, change_x, change_y)
    wrong(dist4, x, y, collated_distances, change_x, change_y)
    wrong(dist5, x, y, collated_distances, change_x, change_y)
    wrong(dist6, x, y, collated_distances, change_x, change_y)
    wrong(dist7, x, y, collated_distances, change_x, change_y)
    wrong(dist8, x, y, collated_distances, change_x, change_y)
    wrong(dist9, x, y, collated_distances, change_x, change_y)

  a =  ( ( max(variance_arr) - min(variance_arr) ) / max(variance_arr) )*100
  b =  ( ( max(bold_arr) - min(bold_arr) ) / max(bold_arr) )*100
  c =  ( ( max(carl_arr) - min(carl_arr) ) / min(carl_arr) )*100
  d =  ( ( max(sil_arr) - min(sil_arr) ) / min(sil_arr) )*100
  e = (1000 - len(x))

  curr_variance.append(a)
  curr_boldmin.append(b)
  curr_carl.append(c)
  curr_sil.append(d)
  curr_anom.append(e)

averagenew(curr_variance, overall_variance)
averagenew(curr_boldmin, overall_boldmin)
averagenew(curr_carl, overall_carl)
averagenew(curr_sil, overall_sil)
averagenew(curr_anom, overall_anomalies)
```
